```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv("./data_twitter/tweets_train.csv")
print data.shape
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x="label",data=data,palette="Blues_d")
plt.ylabel("Frequency",fontsize=14)
plt.xlabel("Class",fontsize=14)
plt.title("Dataset classes distribution",fontsize=20)
plt.show()
```
